```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 2000

data = pd.DataFrame({
    "house_age": np.random.randint(1, 50, n),
    "location_risk": np.random.uniform(0, 1, n),
    "roof_type": np.random.randint(1, 4, n),
    "past_claims": np.random.randint(0, 5, n),
    "property_value": np.random.randint(100000, 1000000, n)
})

# STRONGER LOGIC
data["risk"] = (
    0.5 * data["location_risk"] +
    0.3 * (data["past_claims"] / 5) +
    0.2 * (data["house_age"] / 50)
)

data["risk"] = data["risk"].clip(0, 1)

data.to_csv(r"C:\Users\DELL\Desktop\sample_project\DSMLHINPROJECT\insurance-risk-ml\data\insurance_data.csv", index=False)
```
